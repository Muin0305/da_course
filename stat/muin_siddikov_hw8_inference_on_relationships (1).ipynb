### Задача 1
Решите следующие задачи, принимая уровень значимости равным 5%:

У случайной выборки людей разного возраста спросили надевают ли они очки при чтении. Результаты приведены в таблице ниже. Имеется ли взаимосвязь между возрастом и ношением очков?
                        надевает очки	не надевает
возраст ≤ 20	          15	            56
20 < возраст ≤ 40	      19	            37
40 < возраст ≤ 60	      34	            35
возраст > 60	          51	            23

```python
import scipy.stats as stats
import numpy as np

data = np.array([[15, 56], [19, 37], [34, 35], [51, 23]])

chi2, p, dof, expected = stats.chi2_contingency(data)

chi2, p
```

**Так как p-значение намного меньше уровня значимости 0.05, мы отвергаем нулевую гипотезу о независимости. Это означает, что между возрастом и ношением очков при чтении существует статистически значимая связь.**

### Задача 2
Случайная выборка студентов была разделена на 3 группы и каждая из них готовилась к сдаче экзамена по одной из 3 методик (A, B, C). В таблице ниже приведены оценки по экзамену студентов каждой группы. Определите имеются ли различия между группами.
Group 1	    Group 2	   Group 3
85	          91	    79
86	          92	    78
88	          93	    88
75	          85	    94
78	          87	    92
94	          84	    85
98	          82	    83
79	          88	    85
71	          95	    82
80	          96	    81

```python
group1 = [85, 86, 88, 75, 78, 94, 98, 79, 71, 80]
group2 = [91, 92, 93, 85, 87, 84, 82, 88, 95, 96]
group3 = [79, 78, 88, 94, 92, 85, 83, 85, 82, 81]

# Проведение однофакторного ANOVA теста
f_stat, p_value = stats.f_oneway(group1, group2, group3)

# Вывод результатов
f_stat, p_value
```

**Так как p-значение больше уровня значимости 0.05, мы не отвергаем нулевую гипотезу. Это означает, что статистически значимых различий между группами не обнаружено.**
